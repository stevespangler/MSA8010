# Filtering (with)

### 1. Provide a list of employees that are married.

```python
Employees[['EmployeeID','FirstName','LastName','MaritalStatus']][Employees.MaritalStatus == 'M']
```

### 2a. Show me a list of employees that have a lastname that begins with "R".

```python
# Employees[['EmployeeID','FirstName','LastName']][Employees.LastName.str.startswith('R')]
Employees.loc[Employees.LastName.str.startswith('R'), ['EmployeeID','FirstName','LastName']]
```

### 2b. Show me a list of employees that have a lastname that ends with "r"

```python
Employees[['EmployeeID','FirstName','LastName']][Employees.LastName.str.endswith('r')]
```

### 2c. Provide a list of employees that have a hyphenated lastname.

```python
Employees[['EmployeeID','FirstName','LastName']][Employees.LastName.str.contains('-')]
```

### 3a. Provide a list of employees that are on salary and have more than 35 vacation hours left.

```python
Employees[['EmployeeID','FirstName','LastName','SalariedFlag','VacationHours']][(Employees.SalariedFlag == 1) 
                                                                                & (Employees.VacationHours > 35)]
```

### 3b. Show the same as above but limit it to American employees.

```python
Employees[['EmployeeID','FirstName','LastName','SalariedFlag','VacationHours','CountryName']][(Employees.SalariedFlag == 1) & 
                                                 (Employees.VacationHours > 35) & (Employees.CountryName == 'United States')]
```

### 3c. Show the same as above but limit it to non-American employees.

```python
Employees[['EmployeeID','FirstName','LastName','SalariedFlag','VacationHours','CountryName']][(Employees.SalariedFlag == 1) 
                                                & (Employees.VacationHours > 35) & (Employees.CountryName != 'United States')]
```

### 4a. List the married employees with more than 35 vacation hours, only ones living in Washington state.

```python
Employees[['EmployeeID','FirstName','LastName','MaritalStatus','VacationHours','StateProvinceName']][
    (Employees.MaritalStatus == 'M') & (Employees.VacationHours > 35) & (Employees.StateProvinceName == 'Washington')]
```

### 4b. Change the logic to include anyone who meets any of the 3 conditions (i.e., people who are either married, live in Washington state, or have more than 35 vacation hours left)

```python
Employees[['EmployeeID','FirstName','LastName','MaritalStatus','VacationHours','StateProvinceName']][
    (Employees.MaritalStatus == 'M') | (Employees.VacationHours > 35) | (Employees.StateProvinceName == 'Washington')]
```

### 4c. Show the same as above, but only for Production Technicians

```python
Employees[['EmployeeID','FirstName','LastName','MaritalStatus','VacationHours','StateProvinceName','JobTitle']][
    (Employees.JobTitle.str.contains("Production Technician")) &
    ((Employees.MaritalStatus == 'M') | (Employees.VacationHours > 35) | (Employees.StateProvinceName == 'Washington'))]
```

### 5a. List all employees living in Redmond, Seattle, and Bellevue, showing EmployeeID, FirstName, LastName, and City. Sort the list alphabetically by city.

```python
Employees.sort_values(by='City')[['EmployeeID','FirstName','LastName','City']][
    (Employees.City == 'Redmond')|(Employees.City == 'Seattle')|(Employees.City == 'Bellevue')]
```

### 5b. For the list above, make sure these are only in the state of Washington, just to be careful.

```python
Employees.sort_values(by='City')[['EmployeeID','FirstName','LastName','City']][
    ((Employees.City == 'Redmond')|(Employees.City == 'Seattle')|(Employees.City == 'Bellevue')) 
    & (Employees.StateProvinceName == 'Washington')]
```

### 6. Provide a list of employees who have no title, whether it's a NULL or empty string.

```python
Employees[['EmployeeID','Title','FirstName','LastName']][(Employees.Title=='') | (pd.isnull(Employees.Title))]
```

### 7a. Provide a list of employees who have at least 60 vacation hours left.

```python
Employees[['EmployeeID','FirstName','LastName','VacationHours']][Employees.VacationHours >= 60]
```

### 7b. Provide a list of employees who have less than 60 vacation hours left.

```python
SELECT e.EmployeeID, e.Title, e.FirstName, e.LastName, e.VacationHours, e.SickLeaveHours
	FROM dbo.Employees AS e
	WHERE e.VacationHours < 60
	;
```

```python
Employees[['EmployeeID','FirstName','LastName','VacationHours']][Employees.VacationHours < 60]
```

### 7c. Show me employees who have more than 20 and less than 60 vacation hours left.

```python
Employees[['EmployeeID','FirstName','LastName','VacationHours']][(Employees.VacationHours < 60) & (Employees.VacationHours > 20)]
```

### 7d. If you did not use BETWEEN for 7c, do the same but use BETWEEN. If you did use BETWEEN for 7c, do it another way.

```python
Employees[['EmployeeID','FirstName','LastName','VacationHours']][Employees.VacationHours.between(20,60,inclusive=False)]
```

# Grouping

### 1a. What is the earliest birthdate for all employees?

```python
Employees.BirthDate.min()
```

### 1b. Add to the above, the most recent birthdate for all employees

```python
print('Earliest',Employees.BirthDate.min(),'Most Recent',Employees.BirthDate.max())
```

### 1c. Show the above results broken down by gender

```python
Employees.groupby('Gender')['BirthDate'].agg({'Earliest Birthday':np.min,'Most Recent Birthday':np.max})
```

### 1d. Show the above results broken down by gender, and salaried/hourly

```python
Employees.groupby(['Gender','SalariedFlag'])['BirthDate'].agg({'Earliest Birthday':np.min,'Most Recent Birthday':np.max})
```

### 2a. What are the average vacation hours for all employees?

```python
Employees.VacationHours.mean()
```

### 2b. Add to the above, the minimum vacation hours for all employees

```python
print('Min',Employees.VacationHours.min(),'Mean',Employees.VacationHours.mean())
```

### 2c. Show the above results broken down and ordered by job title

```python
Employees.groupby('JobTitle')['VacationHours'].agg({'Avg Hours':np.mean, 'Min Hours':np.min})
```

### 2d. Show the above results broken down by job title, and married/single employees

```python
Employees.groupby(['JobTitle','MaritalStatus'])['VacationHours'].agg({'Avg Hours':np.mean, 'Min Hours':np.min})
```

### 2e. Add to the above, the maximum vacation hours per group

```python
Employees.groupby(['JobTitle','MaritalStatus'])['VacationHours'].agg({'Avg Hours':np.mean, 'Min Hours':np.min,'Max Hours':np.max})
```

### 2f. Show the above results broken down by job title, married/single employees, and State

```python
Employees.groupby(['JobTitle','MaritalStatus','StateProvinceName'])['VacationHours'].agg({'Avg Hours':np.mean, 'Min Hours':np.min,'Max Hours':np.max})
```

### 2g. Show the above results but only for American employees

```python
Employees[Employees.CountryName == 'United States'].groupby(['JobTitle','MaritalStatus','StateProvinceName','CountryName'])[
    'VacationHours'].agg({'Avg Hours':np.mean, 'Min Hours':np.min,'Max Hours':np.max})
```

### 2h. Change the grouping above so it's broken down by married/single and State, no more job title

```python
Employees[Employees.CountryName == 'United States'].groupby(['MaritalStatus','StateProvinceName'])['VacationHours'].agg(
    {'Avg Hours':np.mean, 'Min Hours':np.min,'Max Hours':np.max})
```

### 2i. Limit the results above to States where the average vacation hours is greater than 30

```python
x = Employees[Employees.CountryName == 'United States'].groupby(['MaritalStatus','StateProvinceName'])['VacationHours'].agg(
    {'AvgHours':np.mean, 'Min Hours':np.min,'Max Hours':np.max})
x[x.AvgHours>30]
```

### 2j. Limit the results above to States where the average vacation hours is greater than 30 and the maximum vacation hours is less than 50

```python
y = Employees[Employees.CountryName == 'United States'].groupby(['MaritalStatus','StateProvinceName'])['VacationHours'].agg(
    {'AvgHours':np.mean, 'MinHours':np.min,'MaxHours':np.max})
y[(y.AvgHours>30) & (y.MaxHours<50)]
```

### 2k. Show the same results but only for non-American employees

```python
z = Employees[Employees.CountryName != 'United States'].groupby(['MaritalStatus','StateProvinceName'])['VacationHours'].agg({'AvgHours':np.mean, 'MinHours':np.min,'MaxHours':np.max})
z[(z.AvgHours>30) & (z.MaxHours<50)]
```

### 4f. Make city and State a single column in the format of "City, State"

```python
res = Employees.groupby(['City','StateProvinceName'])['VacationHours'].agg(
        {'AvgVacHours':np.mean, 'NoEmployees':np.size}).reset_index().sort_values(['City','StateProvinceName'])
res['CityState'] = res.City.str.cat(res.StateProvinceName, sep=', ')
res = res[['City','StateProvinceName','CityState','AvgVacHours','NoEmployees']]
res
```

### 4g. Add a column that shows the difference between the maximum vacation hours and minimum vacation hours for each city

```python
res1 = Employees.groupby(['City','StateProvinceName'])['VacationHours'].agg(
        {'AvgVacHours':np.mean, 'NoEmployees':np.size, 'VacHoursSpread':lambda x: max(x) - min(x)}).reset_index().sort_values(['City','StateProvinceName'])
res1['CityState'] = res1.City.str.cat(res1.StateProvinceName, sep=', ')
res1 = res1[['City','StateProvinceName','CityState','AvgVacHours','VacHoursSpread','NoEmployees']]
res1
```

### 4h. Now sort the results by the new column created above

```python
res1 = Employees.groupby(['City','StateProvinceName'])['VacationHours'].agg(
        {'AvgVacHours':np.mean, 'NoEmployees':np.size, 'VacHoursSpread':lambda x: max(x) - min(x)}).reset_index().sort_values(['City','StateProvinceName'])
res1['CityState'] = res1.City.str.cat(res1.StateProvinceName, sep=', ')
res1 = res1[['City','StateProvinceName','CityState','AvgVacHours','VacHoursSpread','NoEmployees']]
res1.sort_values(['VacHoursSpread'])
```

### 4i. Limit the results to cities that have more than 1 employee

```python
res1 = Employees.groupby(['City','StateProvinceName'])['VacationHours'].agg(
        {'AvgVacHours':np.mean, 'NoEmployees':np.size, 'VacHoursSpread':lambda x: max(x) - min(x)}).reset_index().sort_values(['City','StateProvinceName'])
res1['CityState'] = res1.City.str.cat(res1.StateProvinceName, sep=', ')
res1 = res1[['City','StateProvinceName','CityState','AvgVacHours','VacHoursSpread','NoEmployees']]
res1[res1.NoEmployees > 1].sort_values(['VacHoursSpread'])
```
