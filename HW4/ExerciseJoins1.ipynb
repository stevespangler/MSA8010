### 4d. For each of the items ordered, show the total price of the order (sometimes they ordered more than 1 item)

```python
df = Customers[Customers.StateName == 'Arizona'].merge(Orders, on='CustomerID',how='left')
df['TotalPrice'] = df.Quantity * df.Price
df
```

### 4e. For the Arizona customers above, show the maximum total price per customer. Choose the columns wisely

```python
df.groupby(['CustomerID','LastName','FirstName']).agg({'TotalPrice': max})
```

### 4f. For the results above, only show the ones where the MaxTotalPrice is more than $5.00

```python
df[df.TotalPrice > 5].groupby(['CustomerID','LastName','FirstName']).agg({'TotalPrice': max})
```

### OVERACHIEVER 1: Why do you think a database is designed so the data is split up into different tables?  Why not just put it all in one table?

One large table would result in redundant data if values in columns can take multiple values in other columns. Having multiple tables allows for entities to be assigned certain attributes just once, instead of for every instance in which that entity might appear in relation to other entities.
Also, one large table could quickly become difficult to work with and organize once its size reached a certain level.

### OVERACHIEVER 2a: Take the query from 1b and sort it by the sales region so that all the salespeople are the top of the list

```python
Employees[['EmployeeID','FirstName','LastName','TerritoryID']].merge(Territory[['TerritoryID','Name','CountryCode','Region','SalesYTD','SalesLastYear']],\
         on='TerritoryID', how='left').sort_values('Region')
```

### OVERACHIEVER 2b: From the query above, if the sales territory is outside Europe, do not display it (but we still want to see all the employees in the company!)

```python
Employees[['EmployeeID','FirstName','LastName','TerritoryID']].merge(Territory.loc\
            [(Territory.Region != 'Europe'), ['TerritoryID','Name','CountryCode','Region','SalesYTD','SalesLastYear']],\
         on='TerritoryID', how='left').sort_values('Region')
```
