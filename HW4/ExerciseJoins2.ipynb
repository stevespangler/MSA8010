```python
import pandas as pd
import numpy as np
```

```python
Employees = pd.read_excel('/home/data/AdventureWorks/Employees.xls')
Territory = pd.read_excel('/home/data/AdventureWorks/SalesTerritory.xls')
Customers = pd.read_excel('/home/data/AdventureWorks/Customers.xls')
Orders = pd.read_excel('/home/data/AdventureWorks/ItemsOrdered.xls')
```

### 1a. Provide a count of the number of salespeople in the company

```python
Employees[Employees.JobTitle == 'Sales Representative'].EmployeeID.count()
```

### 1b. Instead of a count, list the salespeople along with their territories
Show columns: EmployeeID, FirstName, LastName, TerritoryName, CountryCode, Region

```python
Employees[['EmployeeID','FirstName','LastName','TerritoryID']].merge(Territory[['TerritoryID','Name','CountryCode','Region']],\
on='TerritoryID', how='inner')
```

### 1c. Now include in the result set all other non-sales employees

```python
Employees.merge(Territory, on='TerritoryID', how='left')[['EmployeeID','FirstName','LastName','Name','CountryCode','Region']]
```

### 1d. Sort the results with the salespeople at the top

```python
Employees.merge(Territory, on='TerritoryID', how='left')[['EmployeeID','FirstName','LastName','Name','CountryCode','Region']]\
.sort_values('Name')
```

### 1e. It turns out we don't need territory information for our American salespeople
Only show information for territories outside the U.S.

```python
Employees.merge(Territory[Territory.CountryCode != "US"], on='TerritoryID', how='left')[['EmployeeID','FirstName','LastName','Name','CountryCode','Region']]\
.sort_values('Name')
```

### 2a. Take the query from #1b above, but add information for each employee's manager
Show additional columns: ManagerFirstName, ManagerLastName, ManagerJobTitle

```python
Employees.columns
```

```python
Employees[['EmployeeID','FirstName','LastName','ManagerID','TerritoryID']]\
.merge(Territory[['TerritoryID','Name','CountryCode','Region']],on='TerritoryID', how='inner')\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
        left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp', '_mgr'))
```

### 2b. Only show salespeople who report to Amy Alberts

```python
df = Employees[['EmployeeID','FirstName','LastName','ManagerID','TerritoryID']]\
.merge(Territory[['TerritoryID','Name','CountryCode','Region']],on='TerritoryID', how='inner')\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
        left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp', '_mgr'))\

df[(df['FirstName_mgr'] == 'Amy') & (df.LastName_mgr == 'Alberts')]
```

### 3a. Give a list of all customers, and the average price of items ordered for each one
Show columns: CustomerID, FirstName, LastName, City, StateName, AvgPrice

```python
Customers.merge(Orders, on='CustomerID', how='left').groupby(['CustomerID','FirstName','LastName','City','StateName'])\
         .Price.agg({'AvgPrice': np.mean})
```

### 3b. For those customers who have never ordered anything, put 0 as the average price

```python
Customers.merge(Orders, on='CustomerID', how='left').groupby(['CustomerID','FirstName','LastName'])\
                .Price.agg({'AvgPrice': np.mean}).fillna(0)
```

### 3c. Show the same results, but instead of average price per customer, tell me the average "Total Price" per customer
Make sure you change the alias of the AvgPrice column accordingly

```python
df = Customers.merge(Orders, on='CustomerID', how='left')
df['TotalPrice'] = df.Quantity * df.Price
df.groupby(['CustomerID','FirstName','LastName']).TotalPrice.agg({'AvgTotPrice':np.mean}).fillna(0)
```

### 4a. Someone has requested a report of all employees and their managers (if they don't have a manager, still display them)
Show columns: EmployeeID, EmployeeFullName, EmployeeTitle, BossFullName, BossTitle

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle','ManagerID']]\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp','_mgr'))
```

### 4b. Another request from someone else has just come in, asking for a list of managers and their employees
Columns: BossFullName, BossTitle, EmployeeFullName, EmployeeTitle
Sort by BossTitle then EmployeeTitle

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle']]\
.merge(Employees[['ManagerID','EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='EmployeeID', right_on='ManagerID', how='inner', suffixes=('_mgr','_emp')).sort(['JobTitle_mgr','JobTitle_emp'])
```

### 4c. Provide a list of managers in the company

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle']]\
.merge(Employees[['ManagerID']],\
       left_on='EmployeeID', right_on='ManagerID', how='inner').drop_duplicates().sort(['LastName','FirstName'])
```

### 5a. Show me only the salespeople who live in the U.S. and their hire dates
Columns: FullName, HireDate

```python
pd.merge(Employees[Employees.CountryName == "United States"], Territory[['TerritoryID']], on='TerritoryID', how='inner')\
[['FirstName','LastName','HireDate']]
```

### 5b. Show me only the salespeople who sell to U.S. customers and their hire dates, plus how much revenue their territory has generated this year
Columns: FullName, HireDate, TerritoryName, Revenue

```python
pd.merge(Employees, Territory[Territory.CountryCode == 'US'][['TerritoryID', 'SalesYTD']], on='TerritoryID', how='inner')\
[['FirstName','LastName','HireDate','SalesYTD']]
```

### 6. I need a list of all the salespeople and their territories, but also show the territories that don't have any salespeople assigned.
Columns: EmployeeID, FirstName, LastName, TerritoryName, TerritoryCountry, TerritoryRegion

```python
pd.merge(Employees[['EmployeeID','FirstName','LastName','TerritoryID']], Territory, on='TerritoryID', how='right').sort('TerritoryID')
```

### 7. Give me a list of all the customers, and if they ordered anything less than $30 show me those orders too

```python
Customers[['FirstName','LastName','CustomerID']].merge(Orders[Orders.Price < 30], on='CustomerID', how='left')
```

### 9a. We need a list of each employee's boss's boss (i.e., 2 levels up)
Columns: EmployeeID, EmployeeFullName, EmployeeTitle, BossBossFullName, BossBossTitle<br/>
Sort by BossBossTitle and then EmployeeTitle

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle','ManagerID']]\
.merge(Employees[['EmployeeID','ManagerID']],\
       left_on='ManagerID', right_on='EmployeeID', how='inner', suffixes=('_emp','_mgr'))\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='ManagerID_mgr', right_on='EmployeeID', how='inner', suffixes=('_emp','_2ndMgr'))\
.sort_values(['JobTitle_2ndMgr','JobTitle_emp'])[['EmployeeID_emp','FirstName_emp','LastName_emp','JobTitle_emp',\
                                                 'FirstName_2ndMgr','LastName_2ndMgr','JobTitle_2ndMgr']]
```

### 9b. Same as above, except show all employees in the company

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle','ManagerID']]\
.merge(Employees[['EmployeeID','ManagerID']],\
       left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp','_mgr'))\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='ManagerID_mgr', right_on='EmployeeID', how='left', suffixes=('_emp','_2ndMgr'))\
.sort_values(['JobTitle_2ndMgr','JobTitle_emp'])[['EmployeeID_emp','FirstName_emp','LastName_emp','JobTitle_emp',\
                                                 'FirstName_2ndMgr','LastName_2ndMgr','JobTitle_2ndMgr']]
```

### 9c. For those employees who don't have a boss's boss, display "n/a" for BossBossFullName and BossBossTitle

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle','ManagerID']]\
.merge(Employees[['EmployeeID','ManagerID']],\
       left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp','_mgr'))\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='ManagerID_mgr', right_on='EmployeeID', how='left', suffixes=('_emp','_2ndMgr'))\
.sort_values(['JobTitle_2ndMgr','JobTitle_emp']).fillna('n/a')[['EmployeeID_emp','FirstName_emp','LastName_emp','JobTitle_emp',\
                                                 'FirstName_2ndMgr','LastName_2ndMgr','JobTitle_2ndMgr']]
```

### 9d. Change the sort so that the ones with "n/a" are at the top of the list

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle','ManagerID']]\
.merge(Employees[['EmployeeID','ManagerID']],\
       left_on='ManagerID', right_on='EmployeeID', how='left', suffixes=('_emp','_mgr'))\
.merge(Employees[['EmployeeID','FirstName','LastName','JobTitle']],\
       left_on='ManagerID_mgr', right_on='EmployeeID', how='left', suffixes=('_emp','_2ndMgr'))\
.sort_values(['JobTitle_2ndMgr','JobTitle_emp'], na_position ='first').fillna('n/a')[['EmployeeID_emp','FirstName_emp','LastName_emp','JobTitle_emp',\
                                                 'FirstName_2ndMgr','LastName_2ndMgr','JobTitle_2ndMgr']]
```

### 10a. We need a list of each manager's indirect reports (i.e., 2 levels down)
Columns: ManagerFullName, ManagerTitle, IndirectReportFullName, IndirectReportTitle<br />
Sort by ManagerTitle and IndirectReportTitle

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle']]\
.merge(Employees[['ManagerID','EmployeeID']],\
       left_on='EmployeeID', right_on='ManagerID', how='inner', suffixes=('_mgr','_emp'))\
.merge(Employees[['ManagerID','FirstName','LastName','JobTitle']],\
      left_on='EmployeeID_mgr', right_on='ManagerID', how='inner', suffixes=('_mgr','_IndRpt'))\
.sort_values(['JobTitle_mgr','JobTitle_IndRpt']).drop_duplicates(['LastName_mgr','LastName_IndRpt'])\
[['FirstName_mgr','LastName_mgr','JobTitle_mgr','FirstName_IndRpt','LastName_IndRpt','JobTitle_IndRpt']]\
```

### 10b. Please tell us how many indirect reports each manager has
Columns: ManagerFullName, ManagerTitle, NumIndirectReports

```python
Employees[['EmployeeID','FirstName','LastName','JobTitle']]\
.merge(Employees[['ManagerID','EmployeeID']],\
       left_on='EmployeeID', right_on='ManagerID', how='inner', suffixes=('_mgr','_emp'))\
.merge(Employees[['ManagerID','EmployeeID','FirstName','LastName','JobTitle']],\
      left_on='EmployeeID_mgr', right_on='ManagerID', how='inner', suffixes=('_mgr','_IndRpt'))\
.sort_values(['JobTitle_mgr','JobTitle_IndRpt']).drop_duplicates(['LastName_mgr','LastName_IndRpt'])\
.groupby(['EmployeeID_mgr','FirstName_mgr','LastName_mgr','JobTitle_mgr']).EmployeeID.agg({'NumIndRpts':np.size})


#[['FirstName_mgr','LastName_mgr','JobTitle_mgr','FirstName_IndRpt','LastName_IndRpt','JobTitle_IndRpt']]
```

### BONUS 1a. The HR people have come to you with a problem, they realized that some of the employee names were entered into the system with 2 spaces instead of 1
